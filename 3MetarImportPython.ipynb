## Halbstunden Merge zu xx:20 und xx:50

```python
import pandas as pd
from traffic.data.weather import metar

# ======================================================
# 1) Einstellungen
# ======================================================
airport_code = "LSZH"  # Z√ºrich Flughafen
START = pd.Timestamp("2023-01-01 05:50", tz="UTC")
STOP  = pd.Timestamp("2025-09-12 23:20", tz="UTC")
STEP  = pd.Timedelta(minutes=30)

OUTPUT_FILE = "metar_official_HALBHOUR_MATCHED.csv"

# ======================================================
# 2) Halbstundenraster erzeugen
# ======================================================
time_grid = pd.date_range(start=START, end=STOP, freq=STEP)
print(f"üïí Erstelle Zeitraster mit {len(time_grid):,} Halbstundenpunkten ...")

# ======================================================
# 3) Offizielle METAR-Daten laden
# ======================================================
print(f"üîÑ Lade offizielle METAR-Daten f√ºr {airport_code} von {START} bis {STOP} ...")

m = metar.METAR
df_metar = m(airport_code).get(start=str(START), stop=str(STOP))

# ======================================================
# 4) Zeitspalte parsen & normalisieren
# ======================================================
df_metar["time"] = pd.to_datetime(df_metar["time"], errors="coerce", utc=True)
df_metar = df_metar.sort_values("time").reset_index(drop=True)

# ======================================================
# 5) Alle Zeitpunkte aus dem Grid beibehalten (auch fehlende)
# ======================================================
# ‚Üí Merge mit vollem Zeitraster, um leere Zeilen f√ºr fehlende METARs zu erzeugen
df_grid = pd.DataFrame({"time": time_grid})
df_full = df_grid.merge(df_metar, on="time", how="left")

# Diagnose: Anzahl fehlender Eintr√§ge pr√ºfen
missing_rows = df_full["raw"].isna().sum() if "raw" in df_full.columns else df_full.isna().all(axis=1).sum()
print(f"‚ö†Ô∏è Fehlende METAR-Zeitpunkte mit leeren Zeilen: {missing_rows}")

# ======================================================
# 6) Speichern
# ======================================================
df_full.to_csv(OUTPUT_FILE, index=False, encoding="utf-8")

print(f"\nüíæ Datei gespeichert als: {OUTPUT_FILE}")
print(f"‚û°Ô∏è Gesamtzeilen (inkl. leere): {len(df_full):,}")
print(f"Zeitraum: {df_full['time'].min()} ‚Äì {df_full['time'].max()}")
print("\nBeispielzeilen:")
print(df_full.head(10))


```

```python
import pandas as pd

# === Datei mit den bereits gefilterten offiziellen METARs (nur vorhandene Zeitpunkte) ===
FILE = "metar_official_HALBHOUR_MATCHED.csv"
# Falls du stattdessen eine andere Datei pr√ºfen willst, einfach den Pfad hier anpassen.

# === Zeitraum & Raster ===
START = pd.Timestamp("2023-01-01 05:50", tz="UTC")
STOP  = pd.Timestamp("2025-09-12 23:20", tz="UTC")
GRID  = pd.date_range(start=START, end=STOP, freq="30min")

# === Datei laden & Zeiten parsen ===
df = pd.read_csv(FILE)
df["time"] = pd.to_datetime(df["time"], errors="coerce", utc=True)

# === Fehlende Zeitpunkte im 30-Minuten-Raster bestimmen (Set-Differenz) ===
metar_times = pd.Index(df["time"].dropna().unique())
missing = sorted(set(GRID) - set(metar_times))

# === Ausgabe ===
print(f"Fehlende Halbstunden-Zeitpunkte: {len(missing)}")
for t in missing:
    print(t)
```
