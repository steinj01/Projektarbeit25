## Finale Tabelle erster Entwurf

```python
import pandas as pd
pd.set_option('display.max_columns', None)  

# ==========================================
# 1) Bestehende Tabellen einlesen
# ==========================================
flights = pd.read_csv("flights_with_weather_vacations_HALBHOUR_merged.csv")
addons = pd.read_csv("addons_merged_HALBHOUR.csv")

# ==========================================
# 2) Sicherstellen, dass beide Tabellen dieselbe Zeilenanzahl und Reihenfolge haben
# ==========================================
print(len(flights), len(addons))  

# ==========================================
# 3) Spalten der addons-Tabelle hinten anhängen
# ==========================================
final = pd.concat([flights, addons], axis=1)

# ==========================================
# 4) Neue Datei speichern (CSV + Pickle)
# ==========================================
final.to_csv("FinaleTabelle.csv", index=False)
final.to_pickle("features.pkl")  

print("✅ Dateien erfolgreich gespeichert:")
print("- FinaleTabelle.csv")
print("- features.pkl")
```
