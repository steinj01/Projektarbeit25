## Erstellen der Faktoren fr die Ferien der einzelnen Kantone

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Liest 'kanton_distanzen_zrh.csv' ein, berechnet:
- Distanzfaktor = (1 - Distanz / max(Distanz))**2
- Einwohnerfaktor = Einwohner / max(Einwohner)
- Gesamtfaktor = Distanzfaktor + Einwohnerfaktor  (max 2)

Und speichert als 'kanton_distanzen_zrh_faktoren.csv'.
Optionales Recompute der Distanz, falls nicht vorhanden (via Haversine).
"""

import pandas as pd
from math import radians, sin, cos, sqrt, atan2
from pathlib import Path
import sys

# Koordinaten des Flughafens Zrich (ZRH)
ZRH_LAT, ZRH_LON = 47.4582, 8.5555


# ------------------------------------------------------------
# Funktion zur Distanzberechnung (Haversine-Formel)
# ------------------------------------------------------------
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Erdumfang in km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c


# ------------------------------------------------------------
# Hauptfunktion
# ------------------------------------------------------------
def main():
    input_path = Path("kanton_distanzen_zrh.csv")
    output_path = Path("kanton_distanzen_zrh_faktoren.csv")

    if not input_path.exists():
        raise FileNotFoundError(f" Datei nicht gefunden: {input_path.resolve()}")

    df = pd.read_csv(input_path)

    # Falls Distanzspalte fehlt  ber Koordinaten berechnen
    if "Distanz_zum_ZRH_km" not in df.columns:
        if not {"Lat", "Lon"}.issubset(df.columns):
            raise ValueError("Weder 'Distanz_zum_ZRH_km' noch 'Lat'/'Lon' vorhanden.")
        df["Distanz_zum_ZRH_km"] = df.apply(
            lambda r: round(haversine(ZRH_LAT, ZRH_LON, float(r["Lat"]), float(r["Lon"])), 1),
            axis=1
        )

    # Prfen, ob Einwohner vorhanden
    if "Einwohner_2025" not in df.columns:
        raise ValueError("Spalte 'Einwohner_2025' fehlt in der Eingabedatei.")

    # ------------------------------------------------------------
    # Faktoren berechnen
    # ------------------------------------------------------------
    max_dist = df["Distanz_zum_ZRH_km"].max()
    max_pop = df["Einwohner_2025"].max()

    df["Distanzfaktor"] = (1 - (df["Distanz_zum_ZRH_km"] / max_dist)) ** 2
    df["Einwohnerfaktor"] = df["Einwohner_2025"] / max_pop
    df["Gesamtfaktor"] = df["Distanzfaktor"] + df["Einwohnerfaktor"]

    # Sortieren nach Distanz (nahe  fern)
    df = df.sort_values("Distanz_zum_ZRH_km").reset_index(drop=True)

    # Speichern
    df.to_csv(output_path, index=False, encoding="utf-8")
    print(f" Datei erfolgreich geschrieben: {output_path.resolve()}")
    print(df[["Kanton", "Distanz_zum_ZRH_km", "Distanzfaktor", "Einwohnerfaktor", "Gesamtfaktor"]].head())


# ------------------------------------------------------------
# Kompatibel mit Jupyter (entfernt berflssige Kernel-Argumente)
# ------------------------------------------------------------
if __name__ == "__main__":
    sys.argv = [sys.argv[0]]
    main()
```

## Erstellen des Gesamtfaktors

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import pandas as pd
from pathlib import Path
import numpy as np
import sys

# ------------------------------------------------------------
# Einstellungen: Dateipfade
# ------------------------------------------------------------
FACTORS_CSV   = "kanton_distanzen_zrh_faktoren.csv"   # enthlt 'Kanton' und 'Gesamtfaktor'
VACATIONS_CSV = "ch_school_vacations_30min.csv"      # DEIN Dateiname hier anpassen (Tippfehler? -> vacations)
OUTPUT_CSV    = "ch_school_vacations_30min_with_index.csv"

# ------------------------------------------------------------
# Hilfen: Kantons-Mapping (Abk. -> Vollname)
# ------------------------------------------------------------
ABBR_TO_NAME = {
    "ZH": "Zrich",
    "BE": "Bern",
    "LU": "Luzern",
    "UR": "Uri",
    "SZ": "Schwyz",
    "OW": "Obwalden",
    "NW": "Nidwalden",
    "GL": "Glarus",
    "ZG": "Zug",
    "FR": "Freiburg",
    "SO": "Solothurn",
    "BS": "Basel-Stadt",
    "BL": "Basel-Landschaft",
    "SH": "Schaffhausen",
    "AR": "Appenzell A.Rh.",
    "AI": "Appenzell I.Rh.",
    "SG": "St. Gallen",
    "GR": "Graubnden",
    "AG": "Aargau",
    "TG": "Thurgau",
    "TI": "Tessin",
    "VD": "Waadt",
    "VS": "Wallis",
    "NE": "Neuenburg",
    "GE": "Genf",
    "JU": "Jura",
}

# Alternative Schreibweisen -> Normalisierung auf Vollnamen
ALT_TO_NAME = {
    "Zurich": "Zrich",
    "Geneva": "Genf",
    "Vaud": "Waadt",
    "Valais": "Wallis",
    "Neuchatel": "Neuenburg",
    "St. Gallen": "St. Gallen",
    "Sion": "Wallis",  # falls flschlich nach Hauptort benannt
}

# ------------------------------------------------------------
# 1) Faktoren laden (Kanton -> Gesamtfaktor)
# ------------------------------------------------------------
factors_df = pd.read_csv(FACTORS_CSV)
if not {"Kanton", "Gesamtfaktor"}.issubset(factors_df.columns):
    raise ValueError("In der Faktoren-Datei fehlen Spalten 'Kanton' oder 'Gesamtfaktor'.")

# Kanton-Namen als Schlssel (sauber getrimmt)
factors_df["Kanton"] = factors_df["Kanton"].astype(str).str.strip()
factors = factors_df.set_index("Kanton")["Gesamtfaktor"]

# ------------------------------------------------------------
# 2) Ferien-Tabelle laden
#    Erwartung: eine Zeitspalte (z.B. 'ts') + pro Kanton eine Bool-Spalte
# ------------------------------------------------------------
vac = pd.read_csv(VACATIONS_CSV)

# Hilfsliste: mgliche Kantonsspalten
cols = list(vac.columns)

# Kandidaten: Abkrzungen
abbr_cols = [c for c in cols if c in ABBR_TO_NAME]
# Kandidaten: Vollnamen
name_cols = [c for c in cols if c in factors.index]

# Falls alternative Schreibweisen auftauchen -> auf Vollnamen mappen
alt_cols = [c for c in cols if c in ALT_TO_NAME]

# Build Mapping von Ferien-Spalten -> Vollname
col_to_fullname = {}

# 2a) Abkrzungen mappen
for c in abbr_cols:
    col_to_fullname[c] = ABBR_TO_NAME[c]

# 2b) Vollnamen direkt bernehmen
for c in name_cols:
    col_to_fullname[c] = c

# 2c) Alternative Schreibweisen
for c in alt_cols:
    target = ALT_TO_NAME[c]
    if target in factors.index:
        col_to_fullname[c] = target

# Falls keine bereinstimmungen: hilfreiche Meldung
if not col_to_fullname:
    raise ValueError(
        "In der Ferien-Tabelle wurden keine Kantonsspalten erkannt.\n"
        "Erwarte entweder Abkrzungen (ZH, BE, ...) oder Vollnamen (Zrich, Bern, ...).\n"
        f"Vorhandene Spalten: {cols}"
    )

# Subset der Ferien-Indikator-Spalten
vac_ind = vac[list(col_to_fullname.keys())].copy()

# Alles robust in 0/1 umwandeln: True/False, 1/0, 'true'/'false', 'yes'/'no' etc.
def to01(x):
    if pd.isna(x):
        return 0
    if isinstance(x, (int, float, np.integer, np.floating)):
        return int(x != 0)
    s = str(x).strip().lower()
    return int(s in {"1", "true", "t", "yes", "y", "ja"})

vac_ind = vac_ind.applymap(to01)

# Spalten auf Vollnamen umbenennen
vac_ind = vac_ind.rename(columns=col_to_fullname)

# Auf die Kantone einschrnken, fr die wir Faktoren haben
available = [c for c in vac_ind.columns if c in factors.index]
if not available:
    raise ValueError("Keine der erkannten Kantonsspalten ist in der Faktoren-Tabelle vorhanden.")

vac_ind = vac_ind[available]

# Reihung der Spalten wie im Faktoren-Index (soweit vorhanden)
order = [k for k in factors.index if k in vac_ind.columns]
vac_ind = vac_ind.reindex(columns=order)

# ------------------------------------------------------------
# 3) Ferien-Gesamtfaktor je 30-Min-Intervall berechnen
#    (Zeilenweiser Skalarprodukt: Ferien-Indicator * Gesamtfaktor)
# ------------------------------------------------------------
# factors auf dieselbe Spaltenreihenfolge bringen
weights = factors.reindex(vac_ind.columns).fillna(0.0).to_numpy()
indicator_matrix = vac_ind.to_numpy(dtype=float)

# Summe der aktiven Kantonsfaktoren je Zeile
sum_factors = (indicator_matrix * weights).sum(axis=1)

# ------------------------------------------------------------
# 4) Neue Spalte ganz am ENDE der Ferien-Tabelle anhngen und speichern
# ------------------------------------------------------------
vac_with_idx = vac.copy()
vac_with_idx["Ferien_Gesamtfaktor"] = sum_factors  # pandas hngt standardmig hinten an

vac_with_idx.to_csv(OUTPUT_CSV, index=False, encoding="utf-8")

print(" Fertig.")
print(f"- Erkannte Kantonsspalten: {list(col_to_fullname.keys())}")
print(f"- Gemappt auf Vollnamen   : {list(vac_ind.columns)}")
print(f"- Ausgabedatei            : {Path(OUTPUT_CSV).resolve()}")
print(vac_with_idx.head())
```

## Entfernen der doppelten Flge

```python
import pandas as pd

# CSV einlesen
df = pd.read_csv("FinaleTabelleMitFerien.csv")

# Zwei neue Spalten aus der nchsten Zeile erzeugen
df["n_arrivals2"] = df["n_arrivals"].shift(-1)
df["n_departures2"] = df["n_departures"].shift(-1)

df["n_arrivals3"] = df["n_arrivals"].shift(-2)
df["n_departures3"] = df["n_departures"].shift(-2)

# Letzte Zeile ohne Targets entfernen
df = df.dropna(subset=["n_arrivals2", "n_departures2"])
df = df.dropna(subset=["n_arrivals3", "n_departures3"])

# Neu: wieder auf int casten
df[["n_arrivals2", "n_departures2"]] = df[["n_arrivals2", "n_departures2"]].astype(int)
df[["n_arrivals3", "n_departures3"]] = df[["n_arrivals3", "n_departures3"]].astype(int)
df["dominant_runway_concept"] = df["dominant_runway_concept"].astype("category")
df["vis_cat"] = df["vis_cat"].astype("category")


# --- NEU: Nachtstunden entfernen (01:20 bis 05:20) ---
df["ts"] = pd.to_datetime(df["ts"], utc=True)

# Anzahl vor dem Lschen merken
rows_before = len(df)

start_night = df["ts"].dt.normalize() + pd.Timedelta(hours=0, minutes=20)
end_night   = df["ts"].dt.normalize() + pd.Timedelta(hours=5, minutes=20)

df = df[~df["ts"].between(start_night, end_night)]
df = df.reset_index(drop=True)

# Anzahl gelschter Zeilen ausgeben
print("Gelschte Zeilen:", rows_before - len(df))
# ----------------------------------------------------

# Speichern
df.to_csv("FinaleTabelle_With_Shifts.csv", index=False)
df.to_pickle("FinaleTabelle_With_Shifts.pkl")
```

## Darstellen der Movements

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# ======================================================
# Style (wie im anderen Programm)
# ======================================================
COLOR_ACTUAL = "#005BAB"   # ZHAW blue (Histogram)

plt.rcParams.update({
    "font.size": 12,
    "axes.titlesize": 13,
    "axes.labelsize": 12,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
})

# ======================================================
# 1) Pfade und Datei
# ======================================================
data_path = Path("FinaleTabelle_With_Shifts.csv")

plot_dir = Path(
    r"C:\Users\janni\OneDrive - ZHAW\5. Semester\Projektarbeit\Programm\Arbeit\Plots"
)
plot_name = "distribution_total_movements_per_30min_interval.pdf"
plot_path = plot_dir / plot_name

# ======================================================
# 2) Daten laden + n_total bauen
# ======================================================
df = pd.read_csv(data_path)

df["n_total"] = pd.to_numeric(df["n_arrivals"], errors="coerce").fillna(0) + \
                pd.to_numeric(df["n_departures"], errors="coerce").fillna(0)

df["n_total"] = df["n_total"].astype(int)

# ======================================================
# 3) Histogramm erstellen (Style angepasst)
# ======================================================
plt.figure(figsize=(8, 5))

plt.hist(
    df["n_total"],
    bins=range(0, df["n_total"].max() + 2),
    color=COLOR_ACTUAL,
    edgecolor="black",
    linewidth=0.4,
    alpha=1.0
)

plt.title("Histogram of Number of Movements per 30-Minute Interval")
plt.xlabel("Number of Movements per Interval (n_total)")
plt.ylabel("Frequency (Number of Intervals)")
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()

# ======================================================
# 4) Speichern
# ======================================================
plt.savefig(plot_path, bbox_inches="tight")
print(f"Plot gespeichert unter: {plot_path}")
plt.show()
```
